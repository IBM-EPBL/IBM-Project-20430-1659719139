## 6. Create a 3x3 matrix with values ranging from 0 to 8

```python
import random
a=[[0]*3]*3
for i in range(3):
  for j in range(3):
    a[i][j]=random.randint(0,9)
a
```

## 7. Concatinate a and b 
## a = np.array([1, 2, 3]), b = np.array([4, 5, 6])

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
final = np.concatenate((a, b), axis = None)
  
print (final)
```
